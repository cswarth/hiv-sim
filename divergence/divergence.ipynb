# Divergence of inferred ancestral sequence from actual founder

Simulated transmission hierarchies are made up of a single donor patient and a single recipient.  Time in the simulation is measured in discrete generations, and the population size is kept constant.   

The donor lineage is evolved for a total of 10,000 generations.  A single sequence is taken from the donor at various points along it's lineage to be the founder of an independent recipient lineage.  Founder sequences might be sampled at generations 100, 300, 500, 1000, 5000, for example.

We are intested in the question of how founder sequence inference is impacted by samples taken from related but independently evolving lineages.   In our simulation we are interested in inferring the founder of the recipient lineage, and the donor is an independently evolving related lineage.   We infer a founder sequence at the MRCA of samples taken from the recipient and donor lineages.  Calculating the distance of an inferred founder from the known founder sequence gives an objective measure to assess the relative impact of differing donor samples. 

Samples of sequences (N=10) are taken from both the donor and recipient lineages and used to construct a phylogenetic tree in BEAST.  Founder sequences are reconstructed at the MRCA of the donor and recipient monophyletic clades.
The combination of donor sample time, donor transmission time, and recipient sample time make one point in our combinatorial space of trees.    Donor sample time is expressed as generations since inception, although for our purposes it is the time since transmission that is important.  The transmission time is expressed as a timepoint along the donor lineage.  Recipient sample time is expressed as time since founding the recipient lineage, which must by definition postdate the transmission event.  Care must be taken to normalize these dates onto a single scale when constructing the phylogenetic analysis.

Plot distance of inferred common ancestor from actual founder as a function of time since transmission of patient-1 for patient-2 time-since-transmission == 700 generations.

```python
%cd ~/src/matsen/hiv-sim/divergence
import os.path

import glob
from Bio import SeqIO
import csv
import pandas as pd
from itertools import ifilter
import matplotlib as plt

def distancesByFile(files):
    df = pd.DataFrame()
    for fname in files:
        tmp = pd.read_csv(fname, names=[fname])
        df = pd.concat([df, tmp], axis=1)
    return(df)

def distance_plot(transmission, recipient):
    glob = "../sims/runs/{}/*/{}/relaxed/dist_detail.csv".format(transmission, recipient)
    files = !ls $glob
    df = distancesByFile(files)

    # set column names to the donor time since transmission
    # print(list(df.columns))
    # print([fname.split("/")[4] for fname in list(df.columns)])
    df.columns = [fname.split("/")[4] for fname in list(df.columns)]
    # sort columns in ascending order
    df = df.reindex_axis(sorted(df.columns, key=lambda item: (int(item)
                               if item[0].isdigit() else -float('inf'), item)), axis=1)
    byDonor = df;
    byDonor.tail(1000).plot(kind='box', figsize=(18,7))
    title = ( "Percent difference of inferred recipient founder from true founder\n"
              "Sampled {} generations post-transmission"
              ).format(recipient)
    plt.pyplot.title(title)
    plt.pyplot.xlabel('sampled donor generation, post-transmission')
    plt.pyplot.ylabel('percentage difference from founder')

plt.style.use('ggplot')
print(matplotlib.__version__)
#execfile("divergence.py")
```

### Plot distance to founder as a function of donor sample time after transmission.

```python

transmission = 300
glob = "../sims/runs/{}/*/*/relaxed/dist_detail.csv".format(transmission)
files = !ls $glob
recipients = sorted(set([fname.split("/")[5] for fname in files]), key=lambda k: int(k))

for r in recipients:
    distance_plot(transmission,r)    
```

```python

transmission = 700
glob = "../sims/runs/{}/*/*/relaxed/dist_detail.csv".format(transmission)
files = !ls $glob
recipients = sorted(set([fname.split("/")[5] for fname in files]), key=lambda k: int(k))

for r in recipients:
    distance_plot(transmission,r)    
```
