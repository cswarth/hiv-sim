```python
%cd ~/src/matsen/hiv-sim/divergence
import os.path

import glob
from Bio import SeqIO
import csv
import pandas as pd
from itertools import ifilter
import matplotlib.pyplot as plt

def versiontuple(v):
    return tuple(map(int, (v.split("."))))

if versiontuple(matplotlib.__version__) < versiontuple("1.4.0"):
    warn("Require matplotlib > v1.4 to use ggplot style")
else: 
    plt.style.use('ggplot')

def distancesByFile(files):
    df = pd.DataFrame()
    for fname in files:
        tmp = pd.read_csv(fname, names=[fname])
        df = pd.concat([df, tmp], axis=1)
    return(df)
    
#execfile("divergence.py")
```

```python

#   replicate/transmissions/donor/recipient/releaxed

glob = "../sims/replicates/*/*/*/*/relaxed/dist_detail.csv"
files = !ls $glob
dist_replicates = distancesByFile(files)
print("\n".join(dist_replicates.columns[1:5] + ["..."]))

def helper(c):
    a = c.split("/")[3:7]
    a.insert(1, '2000')
    return(tuple(a))
tuples = [helper(c) for c in list(dist_replicates.columns)]
print("\n".join([str(t) for t in tuples[1:5]]))

index = pd.MultiIndex.from_tuples(tuples, names=['replicate', 'population', 'transmission','donor','recipient'])
df_p2000 = dist_replicates.copy()
df_p2000.columns = index
```

```python

#   replicate/population/transmission/donor/recipient/releaxed

glob = "../sims/population/*/*/*/*/*/relaxed/dist_detail.csv"
files = !ls $glob
print("\n".join(files[1:5] + ["..."]))
df_distByFile = distancesByFile(files)
tuples = [tuple(c.split("/")[3:8]) for c in list(df_distByFile.columns)]
print("\n".join([str(t) for t in tuples[1:5]]))
index = pd.MultiIndex.from_tuples(tuples, names=['replicate', 'population', 'transmission','donor','recipient'])
df_population = df_distByFile.copy()
df_population.columns = index
```

```python
df = df_population.join(df_p2000)
print(df.head())
```

```python
import matplotlib.lines as mlines

# http://stackoverflow.com/a/4382138/1135316
# http://jsfiddle.net/k8NC2/1/  
colors = [    # boynton colors that are almost never confused
'#0000FF', 
'#FF0000', 
'#00FF00', 
'#FFFF00', 
'#FF00FF', 
'#FF8080', 
'#808080', 
'#800000', 
'#FF8000'

]
showit = True
all_populations = list(set(df.columns.get_level_values('population')))
print(all_populations)
for key, tmp in df.groupby(level=['transmission', 'recipient'], axis=1):

    transmission,recipient = key
    plt.figure();
    populations = list(set(tmp.columns.get_level_values('population')))
    print(populations)
    for replicate, tmp in tmp.groupby(level=['replicate'], axis=1):
        for population, tmp in tmp.groupby(level=['population'], axis=1):
            tmp = tmp.tail(1000).mean()
            tmp.index = sorted(tmp.index.get_level_values('donor'), key=lambda item: (int(item)
                               if item[0].isdigit() else -float('inf'), item))
            if showit:
                print(replicate)
                print(tmp.shape)
                print(tmp.head())
            col = colors[all_populations.index(population)]
            
            tmp.plot(c=col, figsize=(10,6))
            if showit:
                showit = not showit
    title = "Transmission time {}\nRecipient sampled {} generations post-transmission".format(transmission, recipient)
    plt.title(title)
    plt.xlabel('sampled donor generation, post-transmission')
    plt.ylabel('mean % difference from founder')
    marks = [mlines.Line2D([], [], color=colors[all_populations.index(p)], marker=None,
                          label=p) for p in sorted(populations, key=lambda k: int(k))]
    plt.legend(loc='best', title="Population size", handles=marks)


```
