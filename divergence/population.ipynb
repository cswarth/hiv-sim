# Simulated HIV lineages organized by population size
Population size refers to the number of virus genomes independently evolved in the SANTA simulation.   This is actually the carrying capacity, but the simulated population size quickly converges to to this value.

We simulated a full-length HIV-POL gene (2739nt) using four population sizes: 50, 500, 1000, and 2000.

For each population size, four replicates are run.  Each replicate simulates a transmission hierachy where a donor contributes a single founder sequence to a recipient.  The donor and recipient lineages are sampled (N=10, each) at various timepoints at or after the transmission event and BEAST is used to infer a founder sequence at the MRCA of both samples.   BEAST is run for 1,000,000 iterations and founder sequence are logged every 400 iterations.

The charts below show the mean p-distance (or average number of subsitutions) between the actual founder and the inferred founder at each of the last 1000 points in an BEAST simulation.  Only the last 1000 founders are selected to avoid the volatile burnin period.

```python
%cd ~/src/matsen/hiv-sim/divergence
import os.path

import glob
from Bio import SeqIO
import csv
import pandas as pd
import numpy as np
from itertools import ifilter
import matplotlib.pyplot as plt

def versiontuple(v):
    return tuple(map(int, (v.split("."))))

if versiontuple(matplotlib.__version__) < versiontuple("1.4.0"):
    warn("Require matplotlib > v1.4 to use ggplot style")
else: 
    plt.style.use('ggplot')

def distancesByFile(files):
    df = pd.DataFrame()
    for fname in files:
        tmp = pd.read_csv(fname, names=[fname])
        df = pd.concat([df, tmp], axis=1)
    return(df)
    
#execfile("divergence.py")
```

```python

glob = "../sims/population/*/*/*/*/*/relaxed/dist_detail.csv"
files = !ls $glob
df_distByFile = distancesByFile(files)
tuples = [tuple(c.split("/")[3:8]) for c in list(df_distByFile.columns)]
index = pd.MultiIndex.from_tuples(tuples, names=['replicate', 'population', 'transmission','donor','recipient'])
df_population = df_distByFile.copy()
df_population.columns = index
```

```python
import matplotlib.lines as mlines

df = df_population.copy()
# colors to be used to distinguish different population sizes.
# list of boynton colors that are almost never confused
# http://stackoverflow.com/a/4382138/1135316
# http://jsfiddle.net/k8NC2/1/  
colors = [    
'#0000FF', '#FF0000', '#00FF00', '#FFFF00', '#FF00FF', '#FF8080', '#808080', '#800000', '#FF8000'
]
all_populations = list(set(df.columns.get_level_values('population')))


# For each dataset, calculate the average distance to founder for oly the last 1000 inferred
# founder sequences.   This skips the volatile burning period of the MCMC.
df = df.tail(1000).mean()

for transmission, byTransmission in df.groupby(level=['transmission']):
    
    # Two bugs in Pandas make this significantly more complex:
    #    Dataframe.reindex() does not work if the level keyword is supplied
    #
    #    Dataframe.sortlevel() only sorts lexicograpically and does not take a 
    #    key parameter to specify a function to be called on each list element prior to making comparisons
    #
    # As a consequence, we have to use crudy manual indexing here which simply confuses
    # the solution.
    #
    
    recipients = list(unique(byTransmission.index.get_level_values('recipient')))
    recipients.sort(key=lambda item: int(item))
    
    for recipient in recipients:
        # Plot all replicates of a given population size on the same graph
        plt.figure();
        populations = list(set(byTransmission.index.get_level_values('population')))

        byRecipient = byTransmission.loc[byTransmission.index.get_level_values('recipient') == recipient]

        for key,tmp in byRecipient.groupby(level=['population', 'replicate']):  
                population, replicate = key

                keyfunc = lambda item: int(item) if item[0].isdigit() else -float('inf')
                key = map(keyfunc, tmp.index.get_level_values('donor'))
                index = np.argsort(key)
                tmp = tmp.take(index)
                tmp.index = tmp.index.get_level_values('donor')
                
                # choose the color for this line based on population value.
                # Consistently use the same color for a given population on all plots.
                col = colors[all_populations.index(population)]
                tmp.plot(c=col, figsize=(10,6))

        # For each plot, construct a title and legend.
        title = "Transmission time {}\nRecipient sampled {} generations post-transmission".format(transmission, recipient)
        plt.title(title)
        plt.xlabel('sampled donor generation, post-transmission')
        plt.ylabel('mean % difference from founder')
        marks = [mlines.Line2D([], [], color=colors[all_populations.index(p)], marker=None,
                          label=p) for p in sorted(populations, key=lambda k: int(k))]
        plt.legend(loc='best', title="Population size", handles=marks)


```
