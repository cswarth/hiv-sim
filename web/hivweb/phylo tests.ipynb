```python
from cStringIO import StringIO
import collections
import csv
import datetime
import gzip
import itertools
import logging
import os
import os.path
import subprocess
import tempfile

from Bio import SeqIO
from sqlalchemy import func
from sqlalchemy.orm.exc import NoResultFound

from lxml import etree
```

```python
def color_tree(xml_path, nexus_path, show_posterior=True):
    """
    Color query & ignored sequences in a BEAST nexus tree
    """
    doc = etree.parse(xml_path)
    def search_taxa():
        search = []
        taxa = doc.xpath('//superinfectionStatistic/mrca/taxa/@idref')
        if taxa:
            for t in taxa:
                search.extend(doc.xpath('/beast/taxa[@id="{0}"]/taxon/@idref'.format(t)))

        return search

    def ignore_taxa():
        ignore = doc.xpath('//superinfectionStatistic//ignore/taxon/@idref')
        return ignore

    search = search_taxa()
    ignore = ignore_taxa()

    ntf = tempfile.NamedTemporaryFile
    with ntf(prefix='search') as search_handle, \
         ntf(prefix='ignore') as ignore_handle, \
         ntf(prefix='tree_out') as outfile:
        search_handle.write('\n'.join(search))
        search_handle.flush()
        ignore_handle.write('\n'.join(ignore))
        ignore_handle.flush()
        cmd = ['tree-color', nexus_path, outfile.name,
               '0x990000', search_handle.name,
               '0x009900', ignore_handle.name]
        print ' '.join(cmd)
        subprocess.check_call(cmd)
        result = outfile.read()
        if show_posterior:
            result += """
begin figtree;
  set branchLabels.displayAttribute="posterior";
  set branchLabels.significantDigits=2;
  set branchLabels.isShown=true;
end;
"""
        return result

c = color_tree("../sims/runs/300/300/1000/relaxed/beast_in.xml", "../sims/runs/300/300/1000/relaxed/mcc.tree")

    
```
